```python
import sympy

def grad_descent(expr, l1, history , x1 = -5, y1 = 5, iteration = 0):
    x, y = sympy.symbols('x y')
    df_dx = sympy.diff(expr, x)
    df_dy = sympy.diff(expr, y)
    new_x = x1 - l1 * float(df_dx.subs(x, x1))
    new_y = y1 - l1 * float(df_dy.subs(y, y1))

    current_loss = float(expr.subs({x: x1, y: y1}))
    history.append(current_loss)

    if abs(new_x - x1) < 0.001 and abs(new_y - y1) < 0.001:
        return new_x, new_y, iteration
    
    return grad_descent(expr, l1 , history, new_x, new_y, iteration + 1)
```

```python
x, y = sympy.symbols('x y')
loss_history = []
x, y , i = grad_descent((x - 3)**2 + (y + 2)**2, 0.1, loss_history)
print(f"Minimum found at: x = {round(x)}, y = {round(y)} after {i} steps.")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 4))
plt.plot(loss_history)
plt.title('Loss Curve: Descent Progress')
plt.xlabel('Iteration')
plt.ylabel('Function Value f(x, y)')
plt.grid(True)
```
